```python
from __future__ import annotations

import numpy as np
from ttsim.main import main
from ttsim.main_target import MainTarget
from ttsim.main_args import InputData
from ttsim.tt import policy_function, policy_input, param_function, ConsecutiveIntLookupTableParamValue
from ttsim import copy_environment
import mettsim.middle_earth as middle_earth

# Step 1: Create the status quo policy environment
status_quo_environment = main(
    policy_date_str="2000-01-01",
    orig_policy_objects={"root": middle_earth.ROOT_PATH},
    main_target=MainTarget.policy_environment,
)

# Step 2: Create a copy of the status quo environment  
wealth_adjusted_bounty_pe = copy_environment(status_quo_environment)

# Step 3: Create the new policy function (following GETTSIM tutorial pattern exactly)
@policy_function(
    start_date="2000-01-01",
    leaf_name="amount",
)
def amount_modified(
    amount_without_topup: float,
    bounty_topup_by_age: ConsecutiveIntLookupTableParamValue,
    wealth: float,
    age: int,
) -> float:
    """Orc-hunting bounty with wealth adjustment."""
    if wealth > 1000.0:
        # If wealth is above 1000, no topup is applied.
        return amount_without_topup
    else:
        return amount_without_topup * bounty_topup_by_age.look_up(age)

# Step 4: Replace the function in the copied environment
wealth_adjusted_bounty_pe["orc_hunting_bounty"]["amount"] = amount_modified

# Step 5: Generate input data template based on modified policy environment
input_template1 = main(
    main_target=MainTarget.templates.input_data_dtypes,
    policy_date_str="2000-01-01",
    orig_policy_objects={"root": middle_earth.ROOT_PATH},
    policy_environment=wealth_adjusted_bounty_pe,
    tt_targets={"tree": 
        {
            "housing_benefits": {"amount_m_fam": None},
            "orc_hunting_bounty": {"amount": None},
        }
    },
)

input_template1
```

```python
# Outcomment fix in templates.py to get `wealth_fam` in input data
input_template2 = main(
    main_target=MainTarget.templates.input_data_dtypes,
    policy_date_str="2000-01-01",
    orig_policy_objects={"root": middle_earth.ROOT_PATH},
    policy_environment=wealth_adjusted_bounty_pe,
    tt_targets={"tree": 
        {
            "housing_benefits": {"amount_m_fam": None},
            "orc_hunting_bounty": {"amount": None},
        }
    },
    include_warn_nodes=False,
    input_data=InputData.tree(
        tree={
            "p_id": np.array([0, 1, 2]),
            "property_tax": {
                "amount_y": np.array([1000, 0, 0]),
            },
        },
    ),    
)
input_template2
```
